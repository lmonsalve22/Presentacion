## Histogramas

Partamos discutiendo un poco sobre histogramas...

```python
# Veamos que pasa si tratamos de graficar directamente una variable
data['alcohol'].plot();
```

```python
data['alcohol'].plot(kind = 'hist');
```

```python
gdp_clean = data['alcohol'].dropna()


plt.hist(gdp_clean, color = 'dodgerblue')

plt.axvline(gdp_clean.mean(), color = 'tomato', linewidth = 2,
            linestyle = '--')
plt.xlabel('GDP', size = 15)
plt.title('Histograma del alcohol en los vinos', size = 15);
```

```python
sn.distplot(data['alcohol'].dropna());
sn.despine()

plt.title('Distribucin del alcohol en los vinos',fontsize = 18);
```

```python
plt.plot(data['alcohol'].dropna(), data['quality'].dropna(), '+');


plt.title('Alcohol y calidad del vino', fontsize = 16);
plt.xlabel('Alcohol', fontsize = 16)
plt.ylabel('Calidad', fontsize = 16);
```

```python
data.corr()
```

```python
fig, _ = plt.subplots(figsize =(9,6))

sn.scatterplot(x = data['alcohol'].dropna(), y = data['quality'].dropna());

sn.despine()
plt.title('Alcohol y calidad del vino', fontsize = 16);
plt.xlabel('Alcohol', fontsize = 16)
plt.ylabel('Calidad', fontsize = 16);
```
