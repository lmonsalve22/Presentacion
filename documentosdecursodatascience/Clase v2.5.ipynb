## Codificación de variables categóricas

Para poder usar columnas categóricas en un modelo, es necesario transformar las etiquetas de las categorias en alguna representación numérica. Existen varias formas de hacer este proceso, algunas básicas son:

Sea una variable categórica con $K$ clases distintas:

1. **Binary Encoding:** Genera $K-1$ columnas binarias.
2. **One-Hot encoding:** Genera $K$ columnas binarias. Cada columna representa la presencia o ausencia de una de las clases.
3. **Label Encoding:** Genera $1$ sola columna y asiga un numero entero natural a cada clase.

Usaremos One-Hot Encoding, crearemos $5$ columnas, una para cada continente, cuando la columna que representa al continente "Oceania" por ejemplo, tome el valor $1$, quiere decir que el correspondiente registro tiene la etiqueta 'Oceania' en su columna de region.

```python
# Rellenando los valores de la columna
df['Africa'] = df['region'].replace(['Africa', 'Asia', 'Europe', 'Americas', 'Oceania'],
                                     [1,0,0,0,0])

df['Asia'] = df['region'].replace(['Africa', 'Asia', 'Europe', 'Americas', 'Oceania'],
                                     [0,1,0,0,0])

df['Europe'] = df['region'].replace(['Africa', 'Asia', 'Europe', 'Americas', 'Oceania'],
                                     [0,0,1,0,0])

df['Americas'] = df['region'].replace(['Africa', 'Asia', 'Europe', 'Americas', 'Oceania'],
                                     [0,0,0,1,0])

df['Oceania'] = df['region'].replace(['Africa', 'Asia', 'Europe', 'Americas', 'Oceania'],
                                     [0,0,0,0,1])
```

```python
df.sample(5)
```

```python
df['region'].unique()
```

```python
# Podemos hacer lo anterior de forma muchisimo mas rapida con un loop
for reg in df['region'].unique():
    df[reg] = np.where(df['region'] == reg, 1, 0)

df.sample(10)
```

Continuando con el ejercicio, veamos na forma mas limpia de hacer lo anterior y agreguemos una linea vertical que indique la media del `gdp`.

```python
gdp_clean = df['gdp'].dropna()


plt.hist(gdp_clean, color = 'dodgerblue')

plt.axvline(gdp_clean.mean(), color = 'tomato', linewidth = 2,
            linestyle = '--')
plt.xlabel('GDP', size = 15)
plt.title('Histograma del producto interno bruto', size = 15);
```

Implementemos una función simple que repita este proceso:

```python
def plot_hist_custom(var):
    tmp = var.dropna()
    plt.hist(tmp, color = 'dodgerblue')
    plt.axvline(tmp.mean(), color = 'tomato', linewidth = 2,
                linestyle = '--')
```

```python
plot_hist_custom(df['literacy']);
```

```python
plot_hist_custom(df['adfert']);
```

Podemos hacer varios graficos al mismo tiempo en una grilla:

```python
plt.subplot(1,2,1)
plot_hist_custom(df['literacy'])
plt.title('Literacy')

plt.subplot(1,2,2)
plot_hist_custom(df['adfert'])
plt.title('Adfert');
```

<i><h2 style="font-family:serif;font-size:250%; text-align:center;color:#283A5B"> Modelos de Regresión Lineal</h2></i>
<br>

Utilizaremos una aproximación de OLS (Ordinary Least Squares), nuestro modelo tiene la siguiente forma:

$$
y_i = \beta_j x_i + \beta_{j+1} x_{i+1} + ...+ \beta_{n} x_{n} 
$$

OLS minimiza el cuadrado de la distancia entre el punto que predice el modelo y el verdadero.

```python
#import statsmodels.api as sm
import statsmodels.formula.api as smf

modelo = smf.ols('literacy ~ life', df_clean)
```

```python
# Ahora 'Entrenamos' el modelo

modelo = modelo.fit()
modelo.summary()
```

```python
from statsmodels.graphics.regressionplots import abline_plot
#modelo = smf.ols('literacy ~ life', df_clean)

fig, ax = plt.subplots()
sn.scatterplot(df_clean['life'], df_clean['literacy'], ax = ax)
abline_plot(model_results=modelo, ax = ax);
```

```python
# Veamos que sucede cuando agregamos mas variables al modelo

modelo_2 = smf.ols('literacy ~ chldmort + life + gdp', df_clean).fit()
modelo_2.summary()
```
